```python
#1.

class Circle:

  def __init__(self , r):
    self.r = r
  
  def Area(self):
    return 3.14 * self.r * self.r
  
  def Perimeter(self):
    return 2 * 3.14 * self.r


c1 = Circle(5)
print(c1.Area())
print(c1.Perimeter())
```

```python
#3.

class account:

  def __init__(self , Account_No , Account_holder_name , Amount):
    self.Account_No = Account_No
    self.Account_holder_name = Account_holder_name
    self.Total = 0
    self.Amount  = Amount
    
  def Deposit(self , extra_amount = 0):
    if(extra_amount == 0):
      print("You Deposited : " , self.Amount)
    else:
      print("You Deposited : " , extra_amount)
      
    self.Total = self.Total + self.Amount + extra_amount
    print("Your Total Balance : " , self.Total)

  def Withdraw(self):
    if(self.Amount > self.Total):
      print("You donot have enouhg balance \n your balance is : " , self.Total)
    else:
      print("Amount withdrawing : " , self.Amount)
      self.Total = self.Total - self.Amount
      print("Remaining Balance : " , self.Total)

  def check_balance(self):
    print("Your Balance is : " , self.Total)

  def display_account_details(self):
    print("Account_No : " , self.Account_No , "\nAccount_hoder_name : " , self.Account_holder_name , "\nTotal Amount : " , self.Total)
    


person1 = account(123986 , "name1121" , 50000) 
person1.Deposit()
person1.check_balance()
person1.display_account_details()  
person1.Deposit(20500)
    

```

```python
#7.

class details:

  def __init__(self , Name , Age , rollno):
    self.Name   = Name
    self.Age    = Age
    self.rollno = rollno

  def compare_age(self , ob1):
    if ob1.Age == self.Age:
      print("Age is Equal")
    else:
      print("Age is not Equal")

std1 = details("Rahul" , 21 , 33)
std2 = details("Rohit" , 22 , 74)

std1.compare_age(std2)

std1 = details("Rahul" , 23 , 43)
std2 = details("Rohit" , 23 , 54)

std1.compare_age(std2)
```

```python
#8.

class Student_details:

  Sem       = 1 
  Institute = "IIIT DHARWAD"

  def __init__(self , Name , usn):
    self.Name = Name
    self.usn  = usn

  def show(self):
    print(self.Name, "\n", self.usn)

  @classmethod
  def det(cls):
    print(cls.Sem,"\n",cls.Institute)

  @staticmethod
  def explanation():
    print("This is a student details class which has details of the student studing in IIIT DHARWAD and currently in 1st semister")
    


sd1 = Student_details("name6166" , "20bcs0787")
sd1.show()
sd1.det()
sd1.explanation()
    



    


```
