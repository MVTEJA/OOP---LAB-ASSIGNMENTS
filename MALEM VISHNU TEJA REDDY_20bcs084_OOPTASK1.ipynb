```python
def isleapyear(num):

  if num % 400 == 0 :
    return True
  elif num % 100 == 0:
    return False
  elif num % 4 == 0 :
    return True
  else:
    return False


print(isleapyear(2012))
print(isleapyear(1990))
```

```python
def condition(n):
  if(n%2 == 1):
    print("Weird")
  elif(n%2 == 0  and n >= 2 and n <= 5):
    print("Not Weird")
  elif(n%2 == 0 and n >= 6 and n <= 20):
    print("Weird")
  elif(n%2 == 0 and n > 20):
    print("Not Weird")


condition(19)
condition(24)
condition(8)
condition(4)
  
```

```python
Arrivals = ['Hardik', 'Rahul', 'Virat', 'Rishabh', 'Mahi' , 'Rohit' , 'Rahane']

def fashionably_late(Arrivals , name):
  required_persons = math.floor(len(Arrivals)/2)    
  total_persons    = (len(Arrivals))  

  for t in range(required_persons , total_persons-1):

    if Arrivals[t] == name:
      return "fashionably late"

  return "Not fashionably late"



print(fashionably_late(Arrivals , "Mahi"))
print(fashionably_late(Arrivals , "Rohit"))
print(fashionably_late(Arrivals , "Virat"))
print(fashionably_late(Arrivals , "Rahane"))
```

```python
doc_list = ["The Learn Python Challenge Rohit.","They bought a car." , "bangalore"]


def check(doc_list , word):

  indexes = []

  for i , j in enumerate(doc_list):
    words = j.replace("." , "").replace("," , "").lower().split()
    for t in words:
      if(t == word.lower()):
        indexes.append(i)
  print(indexes)

check(doc_list , "car")
check(doc_list , "Bangalore")
check(doc_list , "Rohit")
```
